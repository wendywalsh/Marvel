```python

import requests
# Import hash libary to generate hash to pass during API call
import hashlib 

# Import time libary to generate unique timestamp variable to pass during API call
import time
```

```python
#hashlib.md5(public_key.encode('utf-8')).hexdigest() # need to convert string varible to byte-string for hashfunction to work
```

```python
# create public key variable to pass during API call
public_key = 'aba1da0f407a38478a127b1cc555ea57'

# create public key variable to pass during API call
private_key = '88bf2a9bb1202b169479373ee87fd692b6df3f9a'

# create ts variable to pass during api call
# use time library to generate current time in unix format
# provides sub second precision, helping to guarentee uniqueness at run time 
ts = time.time() 

# create hash variable to pass during api call
# use .encode(/) function to convert string varibles to byte-string for hashlib function to work
# md5 algorithim is used, called out in Marvel API documentation
# for ts variable, convert to string first as variable type is INT
# hexdigest is required - what is expected by the API. Using normal 'digest()' does not work, format not supported by API
hash = hashlib.md5(str(ts).encode('utf-8') + private_key.encode('utf-8') + public_key.encode('utf-8')).hexdigest()
```

```python
# pull character information for Wolverine, using the character ID

marvel = requests.get("https://gateway.marvel.com:443/v1/public/characters/1009718?ts={}&apikey={}&hash={}".format(ts,public_key,hash))

marvel.json()
```

```python
# pull information about what comicbook series Wolverine has been in, using the character ID

marvel = requests.get("https://gateway.marvel.com:443/v1/public/characters/1009718/series?ts={}&apikey={}&hash={}".format(ts,public_key,hash))

marvel.json()
```

```python
# pull information for characters whos name starts with 'wolver'

marvel = requests.get("https://gateway.marvel.com:443/v1/public/characters?nameStartsWith=wolver&ts={}&apikey={}&hash={}".format(ts,public_key,hash))

marvel.json()
```
